```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
plt.style.use('ggplot')
```

```python
perGame = pd.read_csv('pergame18.csv', header = 0)
perGame.drop_duplicates()
list(perGame)
perGame = perGame.drop(['Rk','Tm','Age','3P','3PA','3P%','2P','2PA','2P%'], axis = 1)
advStats = pd.read_csv('advanced18.csv', header = 0)
advStats.drop_duplicates()
advStats = advStats.drop(['Unnamed: 19','Unnamed: 24','Rk','Pos','Age','G','MP','3PAr'], axis = 1)
#plySals = pd.read_csv('../data/salary-data/fixed-average-salaries/fixed-2017.csv', header = 0)
#plySals = plySals[['name','average salary']]

df_17_18 = pd.merge(perGame, advStats, how ='inner', on='Player')
df_17_18 = df_17_18[df_17_18.duplicated('Player') == False].reset_index(drop=True) # remove bad data
cols = df_17_18.columns.tolist()
#cols = cols[:20] + cols[21:] # move fix to first column
#cols.insert(0,'fix')
cols = cols[:21] + cols[22:] # move tm to first column
cols.insert(1,'Tm')
df_17_18 = df_17_18[cols]
#df_17_18_2 = pd.merge(df_17_18_2, plySals, left_on = 'fix', right_on = 'name', how = 'inner')
#df_13_2 = df_13_2.drop('name', axis = 1)
df_17_18 = df_17_18[df_17_18['G'] > 25].reset_index(drop=True)
df_17_18
```

```python
df_17_18.to_csv('player_17_18.csv', sep = ',', index = False)
```
